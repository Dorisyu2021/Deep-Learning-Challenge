## Optimize the Model


```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python
application_df["STATUS"].value_counts()
```

```python
application_df["SPECIAL_CONSIDERATIONS"].value_counts()
```

STATUS and SPECIAL_CONSIDERATIONS can be dropped since the small % of projects

```python
# Drop the non-beneficial EIN column
application_df=application_df.drop(['EIN'],1)
application_df=application_df.drop(['SPECIAL_CONSIDERATIONS'],1)
application_df=application_df.drop(['STATUS'],1)
application_df.head()
```

```python
# display the number of unique values

application_df.nunique()
```

```python
# Look at NAME value counts for binning
name_count=application_df["NAME"].value_counts()
name_count

#  How many name counts are greater than 5?
name_count[name_count>5]
```

```python
# Determine which values to replace if counts are less than or equal to 5.
names_to_replace=list(name_count[name_count<=5].index)

# Replace in dataframe
for app in names_to_replace:
  application_df["NAME"]=application_df["NAME"].replace(app,"Other")
    
# Check to make sure binning was successful
application_df["NAME"].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_count=application_df["APPLICATION_TYPE"].value_counts()
application_count
```

```python
# Determine which values to replace if application counts are less than 500
Applications_to_replace=list(application_count[application_count<=500].index)


# Replace in dataframe
for A in Applications_to_replace:
  application_df["APPLICATION_TYPE"]=application_df["APPLICATION_TYPE"].replace(A,"Other")
# Check to make sure binning was successful
application_df["APPLICATION_TYPE"].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
cla_count=application_df["CLASSIFICATION"].value_counts()
cla_count
```

```python
# Determine which values to replace if counts are less than 1000
cla_to_replace=list(cla_count[cla_count<=1000].index)


# Replace in dataframe
for C in cla_to_replace:
  application_df["CLASSIFICATION"]=application_df["CLASSIFICATION"].replace(C,"Other")
# Check to make sure binning was successful
application_df["CLASSIFICATION"].value_counts()
```

```python
# Generate our categorical variable lists
a_cat=application_df.dtypes[application_df.dtypes=="object"].index.tolist()
a_cat
```

```python
# perform one-hot encoding on the data
application_dfNEW=pd.get_dummies(application_df)
application_dfNEW.head(5)
```

```python
# Split our preprocessed data into our features and target arrays
X=application_dfNEW.drop(["IS_SUCCESSFUL"],axis='columns').values
y=application_dfNEW["IS_SUCCESSFUL"].values


# Split the preprocessed data into a training and testing dataset
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  100
hidden_nodes_layer2 = 30
hidden_nodes_layer3 = 10

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="sigmoid"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
# Train the model
fit_model=nn.fit(X_train_scaled,y_train,epochs=200)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```
